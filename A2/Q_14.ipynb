```python
import numpy as np
import matplotlib.pyplot as plt

# Define the system of first-order ODEs
def f(t, y):
    dydt = y[1]
    dzdt = 2*y[1]/t + 2*y[0]/(t**2) + np.log(t)
    return np.array([dydt, dzdt])

# Implement Euler's method
def euler(f, t_span, y0, h):
    t_values = np.arange(t_span[0], t_span[1] + h, h)
    n = len(t_values)
    y_values = np.zeros((len(y0), n))
    y_values[:,0] = y0
    for i in range(1, n):
        y_values[:,i] = y_values[:,i-1] + h * f(t_values[i-1], y_values[:,i-1])
    return t_values, y_values

# Define initial conditions and time span
y0 = [1, 0]  # y(1) = 1, z(1) = 0
t_span = [1, 2]
h = 0.001  # Step size

# Solve the system of ODEs using Euler's method
t_values, y_values = euler(f, t_span, y0, h)

# Define the exact solution
def exact_solution(t):
    return 7*t/4 + (t**3/2) * np.log(t) - (3/4)*t**3

# Plot the numerical solution
plt.plot(t_values, y_values[0], label='Numerical Solution for y(t)', color='blue')
plt.plot(t_values, y_values[1], label='Numerical Solution for z(t)', color='red')

# Plot the exact solution
plt.plot(t_values, exact_solution(t_values), label='Exact Solution for y(t)', linestyle='--', color='green')

plt.xlabel('t')
plt.title('Numerical and Exact Solutions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation
def f(t, y, yp):
    return (t**3 * np.log(t) - 2*t*yp + 2*y) / (t**2)

# Implement Euler's method
def euler(f, t_span, y0, yp0, h):
    t_values = np.arange(t_span[0], t_span[1] + h, h)
    n = len(t_values)
    y_values = np.zeros(n)
    yp_values = np.zeros(n)
    y_values[0] = y0
    yp_values[0] = yp0
    for i in range(1, n):
        yp_values[i] = yp_values[i-1] + h * f(t_values[i-1], y_values[i-1], yp_values[i-1])
        y_values[i] = y_values[i-1] + h * yp_values[i-1]
    return t_values, y_values

# Define initial conditions and time span
y0 = 1
yp0 = 0
t_span = [1, 2]
h = 0.001  # Step size

# Solve the initial value problem using Euler's method
t_values, y_values = euler(f, t_span, y0, yp0, h)

# Define the exact solution
def exact_solution(t):
    return (7*t)/4 + ((t**3)/2) * np.log(t) - (3/4)*t**3

# Plot the numerical solution
plt.plot(t_values, y_values, label='Numerical Solution', color='blue')

# Plot the exact solution
plt.plot(t_values, exact_solution(t_values), label='Exact Solution', linestyle='--', color='red')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Numerical and Exact Solutions')
plt.legend()
plt.grid(True)
plt.show()
```
