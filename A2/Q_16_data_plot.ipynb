```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from file
data = np.loadtxt('Q_16_data.txt', skiprows=1)

# Extract columns
t = data[:, 0]
error = data[:, 3]
error_bound = data[:, 4]

# Plot error and error bound
plt.plot(t, error, label='Error')
plt.plot(t, error_bound, label='Error Bound')

plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error and Error Bound vs. t')
plt.legend()
plt.grid(True)
plt.show()
```
