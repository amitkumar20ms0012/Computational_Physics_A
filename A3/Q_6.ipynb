Q. 6. Furier transform of a constant function.

```python
import numpy as np
import numpy.fft as ft
import matplotlib.pyplot as plt

# Define the constant function
def constant_function(x, c):
    return c * np.ones_like(x)

# Define the Fourier transform function
def fourier_transform(x, y):
    return ft.fftshift(ft.fft(y))

# Define parameters
c = 5  # Constant value
n = 1000  # Number of points
x_min = -10  # Minimum value of x
x_max = 10  # Maximum value of x

# Generate x values
x = np.linspace(x_min, x_max, n)

# Generate constant function values
y = constant_function(x, c)

# Compute Fourier transform
ft_y = fourier_transform(x, y)

# Generate frequency values
k = ft.fftshift(ft.fftfreq(n, x[1] - x[0]))

# Plot the constant function
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('Constant Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Plot the Fourier transform
plt.subplot(2, 1, 2)
plt.plot(k, np.abs(ft_y))
plt.title('Fourier Transform')
plt.xlabel('Frequency (k)')
plt.ylabel('|FT(f(x))|')
plt.xlim(-10, 10)  # Limit the x-axis to show frequencies around 0
plt.tight_layout()
plt.show()
```
