```python
# np.convolve defination.
def numerical_convolution(x, g, h):
    # Compute the FFT of g(x) and h(x)
    G = ft.fft(g(x))
    H = ft.fft(h(x))
    
    # Multiply the FFTs in the frequency domain
    G_times_H = G * H
    
    # Apply the inverse FFT to obtain the convolution in the spatial domain
    convolution = np.real(ft.ifft(G_times_H))
    
    return convolution
```

```python
import numpy as np
import numpy.fft as ft
import matplotlib.pyplot as plt

# Define the functions g(x) and h(x)
def g(x):
    return np.exp(-x**2)

def h(x):
    return np.exp(-4*x**2)

# Define the analytical solution for the convolution
def analytical_convolution(x):
    return np.sqrt(np.pi) / np.sqrt(5) * np.exp(-4 * x**2 / 5)

# Define the domain of x values
x_min = -4
x_max = 4
n_points = 2000
x_values = np.linspace(x_min, x_max, n_points)
dx = x_values[1] - x_values[0]

# Compute the convolution numerically using DFT
convolved_numerical = np.convolve(g(x_values), h(x_values), mode='same') * dx

# Compute the analytical convolution
convolved_analytical = analytical_convolution(x_values)

# Plotting the results
plt.plot(x_values, convolved_analytical, label='Analytical Convolution', lw=3, alpha=0.8)
plt.plot(x_values, convolved_numerical, label='Numerical Convolution', lw=3, alpha=0.8)
plt.legend()
plt.title('Comparison of Analytical and Numerical Convolution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
